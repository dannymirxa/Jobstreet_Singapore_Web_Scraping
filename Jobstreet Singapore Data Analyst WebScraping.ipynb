```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#link for Jobstreet Singapore for Data Analyst job with salary above SGD4000
URL = "https://www.jobstreet.com.sg/en/job-search/data-analyst-jobs/?salary=4000&salary-max=2147483647"
```

```python
page = requests.get(URL)
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
```

print(soup.prettify())

```python
joblist = []
```

```python
def extract_from_result(soup): 
    
    for div in soup.find_all("article", attrs={"class": "sx2jih0 sx2jih1 zcydq85u zcydq8h c6ROG_0 _18qlyvc14 _18qlyvc17 zcydq832 zcydq835"}):
        
        company = div.find("span", attrs={"class":"sx2jih0 zcydq84u _18qlyvc0 _18qlyvc1x _18qlyvc1 _18qlyvca"}).text
        
        for locations in div.find_all("span", attrs={"class":"sx2jih0 zcydq84u _18qlyvc0 _18qlyvc1x _18qlyvc3 _18qlyvc7"}):
            locations = div.find("span", attrs={"class":"sx2jih0 zcydq84u zcydq80 iwjz4h0"}).text
        
        #location and salary uses classes with similar name, the different is location contains another class while salary is not
        #code below checks if the class contains a child
        salary = []
        salaries = div.find_all("span", attrs={"class":"sx2jih0 zcydq84u _18qlyvc0 _18qlyvc1x _18qlyvc3 _18qlyvc7"})
        for s in salaries:
            for c in s.children:
                if not hasattr(c,'children'):
                    salary = c
                else:
                    salary = 0
                    
        job = div.find("span", attrs={"class":"sx2jih0"}).text
        
        d = div.find("ul", attrs={"class":"sx2jih0 sx2jih3 h6p8rp0 h6p8rp5"}).text
        
        #for desc in d:
        #    desc = div.find_all("span", attrs={"class":"sx2jih0 zcydq84u _18qlyvc0 _18qlyvc1x _18qlyvc1 _1d0g9qk4 _18qlyvc8"})
        #    for d in desc:
        #        for c in d.children:
        #            desces = c
        
        #desc = div.find_all("span", attrs={"class":"sx2jih0 zcydq84u _18qlyvc0 _18qlyvc1x _18qlyvc1 _1d0g9qk4 _18qlyvc8"})
        #for d in desc:
        #    for c in d.children:
        #        desces = c
                
        jobs = {
            "company" : company,
            "place" : locations,
            "payment" : salary,
            "role" : job,
            "descriptions" : d
        }
        
        joblist.append(jobs)
        
        #print(d)
        
    return
```
